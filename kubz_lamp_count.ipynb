```python
import requests
import json
from collections import Counter
import time
def get_kubz_info(id):
    url = 'https://genesis-api.keungz.com/teaming/token-info'
    headers = {'Content-Type': 'application/json'}
    payload = {
        "kubzTokenIds": id
    }
    data = None
    try:
        response = requests.post(url, headers=headers, json=payload)
        response.raise_for_status()  # 如果请求不成功，会引发异常
        data = response.json()
        #print(data)
    except requests.exceptions.RequestException as e:
        print('Error at id:',id , e)
    return data
```

```python
query = []
for i in range(1,10001):
    query.append(str(i))
data = get_kubz_info(query)
```

```python
def save_dict(dictionary, file_path):
    with open(file_path, 'w') as file:
        json.dump(dictionary, file)


kubz = data['kubz']
save_dict(kubz, 'kubz.json')
```

```python

total_count = {}
for id in range(1,10001):
    lampz_list = list(kubz[str(id)]['items'].values())
    lampz_counter = dict(Counter(lampz_list))
    for key, value in lampz_counter.items():
        if key in total_count:
            total_count[key] += value
        else:
            total_count[key] = value
print(total_count)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.figure(dpi=500)
data = total_count
sorted_data = dict(sorted(data.items(), key=lambda x: x[1]))
labels = list(sorted_data.keys())
values = list(sorted_data.values())
plt.figure(figsize=(10, 6))
bars = plt.barh(labels, values, color='skyblue')
for bar, value in zip(bars, values):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, value, ha='left', va='center')
plt.xlabel('Count')
plt.ylabel('Categories')
plt.title('KUBZ Lampz and Rugz')
plt.tight_layout()
plt.savefig('kubz_lamp_count.svg', format='svg', dpi=300)
plt.show()

```
